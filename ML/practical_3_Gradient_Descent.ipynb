```python
import matplotlib.pyplot as plt
```

```python
def fun(x):
    return (x+3) **2

def grad(x):
    return 2* (x+3)
```

```python
#parameters for gradient descent
x=2                       #initial guess
learning_rate = 0.1       # step size
epochs = 50               # number of iterations

#for plotting 
x_vals = [x]
y_vals = [fun(x)]
```

```python
#gradient descent loop
for i in range(epochs):
    dx=grad(x)
    x=x - learning_rate *dx

    x_vals.append(x)
    y_vals.append(fun(x))

    print(f"iteration {i+1}: x = {x:5f}, f(x) = {fun(x):5f}")
```

```python
print(f"\nLocal minimum at x = {x:5f} f(x) = {fun(x):.5f}")
```

```python
# Plotting the function and steps

x_plot = [i for i in range(-10, 5)]

y_plot = [fun(i) for i in x_plot]

plt.figure(figsize=(8,5))

plt.plot(x_plot, y_plot, label="y = (x + 3)^2")

plt.scatter(x_vals, y_vals, color='red', label="Gradient Descent Steps")

plt.plot(x_vals, y_vals, linestyle='--', color='gray', alpha=0.6)

plt.title("Gradient Descent to Find Local Minima")

plt.xlabel("x")

plt.ylabel("y")

plt.legend()

plt.grid(True)

plt.show()
```
